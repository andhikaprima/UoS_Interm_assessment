```python
import pandas as pd
```

```python
#read data BS1
BS1 = pd.read_excel(r'Xlxs/3_sharks_6Nov20.xlsx', 
                    sheet_name='BS1_ML', 
                    header=None)
BS1.head()
```

```python
# transpose
BS1_transpose = BS1.transpose()
```

```python
# give column name
BS1_transpose.columns = BS1_transpose.iloc[0]
```

```python
# delete rows which already used for column name
BS1_1 = BS1_transpose.drop(BS1_transpose.index[0])
BS1_1.head()
```

```python
#read data BS2
BS2 = pd.read_excel(r'Xlxs/3_sharks_6Nov20.xlsx', 
                    sheet_name='BS2_ML', 
                    header=None)
BS2.head()
```

```python
# transpose
BS2_transpose = BS2.transpose()
```

```python
# give column name
BS2_transpose.columns = BS2_transpose.iloc[0]
```

```python
# delete rows which already used for column name
BS2_1 = BS2_transpose.drop(BS2_transpose.index[0])
BS2_1.head()
```

```python
#get feature only from BS2
BS2_2 = BS2_1.iloc[:, 1:3476]
BS2_2.head()
```

```python
# merge data BS1 and BS2
# Place the DataFrames side by side
data_model = pd.concat([BS1_1, BS2_2], axis=1)
data_model.head()
```

```python
#EXPORT and save data frame to csv
data_model.to_csv(r'Output/6Nov20_combine.csv', 
          index = False, header = False)
```

```python
# Open data model combined
data_model = pd.read_csv(r'Output/6Nov20_combine.csv',sep=",", header=None)
data_model.head()
```

```python
#Train data with Logistic Regression, SVM, Random Forest
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# split data for traning and testing 
X_train, X_test, y_train, y_test = train_test_split(data_model.loc[:, ~data_model.columns.isin([0])], data_model[0],
                                                    test_size=0.30,random_state=8)
```

```python
#create model
lr_mdl = LogisticRegression()
svm_mdl = svm.SVC(C=1.0, kernel='rbf', probability = True)
rf_mdl = RandomForestClassifier(n_estimators=11)
```

```python
model_lr = lr_mdl.fit(X_train,y_train)
model_svm = svm_mdl.fit(X_train,y_train)
model_rf = rf_mdl.fit(X_train,y_train)
```

```python
#predict test data
y_pred_lr = model_lr.predict(X_test)
y_pred_svm = model_svm.predict(X_test)
y_pred_rf = model_rf.predict(X_test)
```

```python
#compute accuracy
print("Logistic Regression Accuracy:",metrics.accuracy_score(y_test, y_pred_lr))
print("SVM Accuracy:",metrics.accuracy_score(y_test, y_pred_svm))
print("Random Forest Accuracy:",metrics.accuracy_score(y_test, y_pred_rf))
```

```python
#Other Metrics
#Create and print Confussion Matix
print("Logistic Regression: \n",metrics.confusion_matrix(y_test, y_pred_lr))
print("SVM: \n",metrics.confusion_matrix(y_test, y_pred_svm))
print("Random Forest: \n",metrics.confusion_matrix(y_test, y_pred_rf))
```

```python
#calculate ROC and draw
#Use predict_proba to get probability of positive

y_pred_prob_lr = model_lr.predict_proba(X_test)
y_pred_prob_svm = model_svm.predict_proba(X_test)
y_pred_prob_rf = model_rf.predict_proba(X_test)

y_prob_lr = y_pred_prob_lr[:, 1]
y_prob_svm = y_pred_prob_svm[:, 1]
y_prob_rf = y_pred_prob_rf[:, 1]
```
