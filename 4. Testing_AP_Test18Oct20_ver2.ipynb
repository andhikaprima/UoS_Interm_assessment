```python
import pandas as pd
```

```python
#read data BS1
BS1 = pd.read_excel(r'Xlxs/2_sharks_18Oct20.xlsx', 
                    sheet_name='BS1', 
                    header=None)
BS1.head()
```

```python
# transpose
BS1_transpose = BS1.transpose()
```

```python
# give column name
BS1_transpose.columns = BS1_transpose.iloc[0]
```

```python
# delete rows which already used for column name
BS1_1 = BS1_transpose.drop(BS1_transpose.index[0])
BS1_1.head()
```

```python
#read data BS2
BS2 = pd.read_excel(r'Xlxs/2_sharks_18Oct20.xlsx', 
                    sheet_name='BS2', 
                    header=None)
BS2.head()
```

```python
# transpose
BS2_transpose = BS2.transpose()
```

```python
# give column name
BS2_transpose.columns = BS2_transpose.iloc[0]
```

```python
# delete rows which already used for column name
BS2_1 = BS2_transpose.drop(BS2_transpose.index[0])
BS2_1.head()
```

```python
#get feature only from BS2
BS2_2 = BS2_1.iloc[:, 1:3476]
BS2_2.head()
```

```python
# merge data BS1 and BS2
# Place the DataFrames side by side
data_model = pd.concat([BS1_1, BS2_2], axis=1)
data_model.head()
```

```python
#EXPORT and save data frame to csv
data_model.to_csv(r'Output/18Oct20_combine.csv', 
          index = False, header = False)
```

```python
# Open data model combined
data_model = pd.read_csv(r'Output/18Oct20_combine.csv',sep=",", header=None)
data_model.head()
```

```python
#get feature and label only
data_model_X = data_model.iloc[:, 1:8153]
data_model_X.head()
```

```python
import pickle
```

```python
# 1. logistic regression standar model
#membuka kembali model untuk digunakan dalam prediksi senyawa baru
with open("Output/model_logistic", "rb") as input_file:
    model_baru = pickle.load(input_file)
```

```python
#menggunakan model tersimpan untuk prediksi
y_MIC_pred_lr = model_baru.predict(data_model_X)
```

```python
y_MIC_pred_lr
```

```python
y_pred_prob_lr = model_baru.predict_proba(data_model_X)
y_prob_lr = y_pred_prob_lr[:, 1]
y_prob_lr
```

```python
prediction = pd.DataFrame({'Actual Species': data_model[0],'Species prediction': y_MIC_pred_lr, 'Probability': y_prob_lr}, columns=['Actual Species', 'Species prediction', 'Probability'])
prediction
```
